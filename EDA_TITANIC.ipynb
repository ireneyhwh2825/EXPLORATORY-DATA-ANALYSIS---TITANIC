**1. IMPORT LIBRARIES**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

**2. LOAD DATASET**

```python
df = pd.read_csv("titanic.csv")

df.head()
```

**3. EXPLORE DATASET**

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

**4. ANALYIZE MISSING VALUES**

```python
df.isnull().sum()
```

```python
#missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
#Text columns
df.fillna("Unknown", inplace=True)
```

**5. ANALYIZE DUPLICATE VALUES**

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

**6. VISUALIZATIONS**

```python
#1. Survival count
sns.countplot(x='Survived', data=df)

plt.title("Passengers Survival Count")

plt.show()
```

```python
# Survival percentage

survival_rate = df['Survived'].mean() * 100

print(f"Survival Rate: {survival_rate:.2f}%")
```

```python
#2. Gender VS Survival
sns.countplot(
    x='Sex',
    hue='Survived',
    data=df
)

plt.title("Gender vs Survival")

plt.show()
```

```python
#3. Passenger class
sns.countplot(
    x='Pclass',
    hue='Survived',
    data=df
)

plt.title("Passenger Class vs Survival")

plt.show()
```

```python
#4. Age distribution
sns.histplot(
    df['Age'],
    bins=20
)

plt.title("Age Distribution")

plt.show()
```

1. The Titanic dataset contains 891 passengers and 12 features.

** Titanic Dataset EDA Report**
-----------------------------------

** Objective**

The objective of this project was to perform Exploratory Data Analysis (EDA) on the Titanic dataset and identify factors that influenced passenger survival.

** Data Cleaning**

The dataset was analyzed for missing and duplicate values. Missing values in numerical columns were replaced using the mean, while missing categorical values were replaced with "Unknown". No duplicate records were found in the dataset.

** Analysis and Findings**

Several visualizations were created to understand the data and identify patterns.

* The survival count analysis showed that more passengers died than survived.
* Female passengers had significantly higher survival rates compared to male passengers.
* First-class passengers had better survival chances than passengers in second and third class.
* Most passengers were between 20 and 40 years old.
* Correlation analysis indicated that passenger class and fare were related to survival outcomes.

** Conclusion**

The analysis revealed that passenger class, gender, and fare were important factors affecting passenger survival. Through data cleaning, visualization, and statistical analysis, meaningful insights were extracted from the dataset, demonstrating the importance of Exploratory Data Analysis in understanding real-world data.
